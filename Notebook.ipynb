```python
X=df[["Average Cost for two","Has Table booking","Has Online delivery","Price range"]]
```

```python
y=df[["Aggregate rating"]]
```

```python
X
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
X["Has Table booking"]=labelencoder.fit_transform(X["Has Table booking"])
```

```python
X
```

```python
X['Has Online delivery']=labelencoder.fit_transform(X['Has Online delivery'])
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np
```

```python
def modeleesult(predictions):
    print("mean_absolute_error on model is {}".format(mean_absolute_error(y_test,predictions)))
    print("root mean square error {}".format(np.sqrt(mean_squared_error(y_test,predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
predictionsfromlr=lr.predict(X_test)
```

```python
modeleesult(predictionsfromlr)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svrmodel=SVR()
```

```python
param_gridsvr={"C":[0.1,0.5],"kernel":["linear","poly"],"degree":[2]}
```

```python
gridsvr=GridSearchCV(svrmodel,param_gridsvr)
gridsvr.fit(X_train,y_train)
```

```python
predsgridsvr=gridsvr.predict(X_test)
```

```python
modeleesult(predsgridsvr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid={
    "max_depth":[10,30],
    "min_samples_leaf":[1,2],
    "min_samples_split":[2,5]
}
```

```python
treemodel=DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator=treemodel,param_grid=param_grid)
```

```python
grid_tree.fit(X_train,y_train)
```

```python
treepredictions=grid_tree.predict(X_test)
```

```python
modeleesult(treepredictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel=RandomForestRegressor()
```

```python
param_gridrfr={"max_depth":[5,15],
               "n_estimators":[2,5,10]}
```

```python
gridrfr= GridSearchCV(rfrmodel,param_gridrfr)
```

```python
gridrfr.fit(X_train,y_train)
```

```python
rfrpredictions = gridrfr.predict(X_test)
```

```python
modeleesult(rfrpredictions)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn_gridparam={"n_neighbors":[9,11,13,15,17]}
knn_grid_search = GridSearchCV(knn,knn_gridparam)
```

```python
knn_grid_search.fit(X_train,y_train)
```

```python
knnpredictions=knn_grid_search.predict(X_test)
```

```python
modeleesult(knnpredictions)
```

```python
knn_grid_search.best_params_
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor()
ada_gridparams={"n_estimators":[50,100,200],
                "learning_rate":[0.1,0.5,1]}
```

```python
ada_gridsearch=GridSearchCV(ada,ada_gridparams)
```

```python
ada_gridsearch.fit(X_train,y_train)
```

```python
adapred=ada_gridsearch.predict(X_test)
```

```python
modeleesult(adapred)
```
